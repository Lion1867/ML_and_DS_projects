```python
import pandas as pd
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from sklearn.model_selection import train_test_split
import torch

# Imports from our package
from lightautoml.automl.presets.tabular_presets import TabularAutoML, TabularUtilizedAutoML
from lightautoml.tasks import Task
```

```python
# для воспроизводимости результатов зафиксируем сид
random_state =18
n_threads = 4 # число потоков, которые будут обрабатывать данные
np.random.seed(random_state)
torch.set_num_threads(n_threads)
```

```python
path = "/content/drive/MyDrive/Datasets/Car_Price/"
```

```python
train = pd.read_csv(path+'train_data.csv')
test = pd.read_csv(path+'test_data.csv')
```

```python
test.head()
```

```python
target_name = 'final_price'
```

```python
train_data, test_data = train_test_split(
    train,
    test_size =0.2,
    random_state = random_state)

train_data
```

#Создание и обучение пресета TabularAutoML
Параметры task
* 'reg' - для задачи регрессии,
* 'binary' - для задачи бинарной классификации,
* 'multiclass' - для задачи многокласссовой классификации


```python
task = Task(
    'reg',
    loss = 'mse',
    metric = 'r2'
)
```

```python
roles = {
    'target': target_name,
    'drop': ['row_ID']
}
```

```python
n_folds = 5
timeout = 60 * 30
```

```python
automl = TabularAutoML(
    task = task,
    timeout = timeout,
    cpu_limit = n_threads,
    reader_params = {'n_jobs': n_threads, 'cv': n_folds, 'random_state': random_state}
)
```

```python
%%time
model_pred = automl.fit_predict(train_data, roles = roles, verbose =1)
```

```python
print (automl.create_model_str_desc())
```

```python
%%time
preds = automl.predict(test_data)
print (f'Предсказанные значения {preds}')
```

```python
preds[0].dtype
```

```python
from IPython.display import display

# Преобразуем preds в одномерный массив
preds_1d = preds.data.flatten()

# Создаем DataFrame из предсказаний
predictions_df = pd.DataFrame(preds_1d, columns=['Predictions'])

# Отображаем DataFrame красиво
display(predictions_df.head(20))
```

```python
final_price_array = np.array(test_data['final_price'])
final_price_array
```

```python
# Создаем список conv_preds из предсказаний
conv_preds = [item[0] for item in preds.data]

# Переводим final_price в numpy массив
final_price_array = np.array(test_data['final_price'])

# Переводим pred_array в numpy массив
preds_array = np.array(conv_preds)
```

## Вывод получившихся значений метрик качества

```python
# Расчитываем MSE
mse = mean_squared_error(final_price_array, preds_array)
print('MSE: ', mse)

# Расчитываем RMSE
rmse = np.sqrt(mse)
print('RMSE: ', rmse)

# Расчитываем MAE
mae = mean_absolute_error(final_price_array, preds_array)
print('MAE: ', mae)

# Расчитываем коэффициент определения R²
r2 = r2_score(final_price_array, preds_array)
print('R²: ', r2)
```
